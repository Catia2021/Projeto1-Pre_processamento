#Passaremos a analisar  se  há  valores  faltantes

```python
CarrosEletricos.isnull()
```

```python
CarrosEletricos.isnull().sum()
```

```python
CarrosEletricos.loc [pd.isnull(CarrosEletricos['TOTAL GERAL'])]
```

**De posse  da análise acima, é  verificado número significativo de valores faltantes.  No sentido de correção destes  valores,  foi  utilizado a  função "fillna" para  colocar o  valor zero nos registros faltantes.**
**E nas linhas  que apresentavam todas  as colunas com   registros faltantes, foi utilizado a  função "dropna" para excluir toda a linha, uma vez  que  tratava-se, apenas,   de uma linha  para separar os estados. **
A confirmação destas alterações são verificadas abaixo.

```python
CarrosEletricos.dropna(how='all', inplace=True)
```

```python
CarrosEletricos.loc [pd.isnull(CarrosEletricos['TOTAL GERAL'])]
```

```python
CarrosEletricos.isnull().sum()
```

```python
CarrosEletricos.fillna(0,inplace=True)
```

```python
CarrosEletricos.isnull().sum()
```
